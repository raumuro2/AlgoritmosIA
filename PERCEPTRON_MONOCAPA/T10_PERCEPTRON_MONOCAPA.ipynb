```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de Entrada:
X1 = np.array([1, 0, 3, 5, 8, 3, 5])
X2 = np.array([1, 2, 3, 7, 9, 8, 6])
D = np.array([0, 0, 1, 1, 1, 1, 1])

# InicializaciÃ³n
w1 = np.random.rand()
w2 = np.random.rand()
b = np.random.rand()
alfa = 0.1

for epocas in range(10):
    # Calculo del Perceptron:
    Z = X1 * w1 + X2 * w2 - b
    Y = np.where(Z > 0, 1, 0)

    # Ajuste de Pesos
    for i in range(len(Z)):
        error = D[i] - Y[i]
        b_inc = -alfa * error
        w1_inc = alfa * error * X1[i]
        w2_inc = alfa * error * X2[i]
        b += b_inc
        w1 += w1_inc
        w2 += w2_inc

    # Plotting
    plt.pause(1)  # Espera un segundo
    f = lambda x: -w1 / w2 * x + b / w2
    plt.plot(X1, X2, '*r')
    x_vals = np.linspace(-10, 10, 400)
    y_vals = f(x_vals)
    plt.plot(x_vals, y_vals)
    plt.axis([-10, 10, -10, 10])
    plt.show()
    plt.clf()

    # Comprobacion del error:
    Z = X1 * w1 + X2 * w2 - b
    Y = np.where(Z > 0, 1, 0)
    E = D - Y
    if abs(sum(E)) < 0.1:
        break

print("Final weights and bias:", w1, w2, b)
```
